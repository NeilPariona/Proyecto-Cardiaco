```python
import pandas as pd

dataset = "https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/refs/heads/master/data/heart-disease.csv"
df = pd.read_csv(dataset)

print("Dataset cargado")
df.head()
```

3. Realice el análisis exploratorio de datos que considere necesario (por ejemplo,
df.info, df.shape, df.describe, etc.). Genere al menos 3 gráficas y explique su
importancia para conocer mejor el dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Información básica
print(df.info())
print(df.shape)
print(df.describe())
```

```python
# Gráfica 1: Distribución de la variable objetivo
plt.figure(figsize=(6,4))
sns.countplot(x='target', data=df)
plt.title('Distribución de enfermedad cardíaca    0=No    1=Sí')
plt.show()
```

```python
# Gráfica 2: Matriz de correlación
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de correlación')
plt.show()
```

```python
# Gráfica 3: Edad vs presión arterial máxima por target
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='age', y='trestbps', hue='target', alpha=0.7)
plt.title('Edad vs Presión arterial en reposo')
plt.xlabel('Edad')
plt.ylabel('Presión arterial en reposo')
plt.show()
```

4. Realice el procesamiento de datos que requiera su dataset. Justifique

```python
from sklearn.preprocessing import StandardScaler

# Separar características (X) y variable objetivo (y)
X = df.drop('target', axis=1)
y = df['target']

# Verificar valores nulos
print(f"Valores nulos por columna:\n{df.isnull().sum()}")

# Escalar características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Procesamiento completado")
print(f"Shape de X_scaled: {X_scaled.shape}")
```

#Parte 2. MODELO

5. Realice la división de datos. Considere usar stratify, validación cruzada u otro de
forma justificada.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Entrenamiento: {X_train.shape[0]} muestras")
print(f"Prueba: {X_test.shape[0]} muestras")
print(f"Proporción en y_train: {y_train.value_counts(normalize=True)}")
```

6. Realice el entrenamiento. Genere al menos dos modelos de machine learning.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Modelo 1: Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)
modelo_rf.fit(X_train, y_train)

# Modelo 2: SVM
modelo_svm = SVC(kernel='rbf', probability=True, random_state=42)
modelo_svm.fit(X_train, y_train)

print("Modelos entrenados, vamo a segui")
```

7. Evalúe al menos dos métricas y coméntelas.

```python
from sklearn.metrics import accuracy_score, precision_score, classification_report

# Predicciones
y_pred_rf = modelo_rf.predict(X_test)
y_pred_svm = modelo_svm.predict(X_test)

# Métricas
acc_rf = accuracy_score(y_test, y_pred_rf)
prec_rf = precision_score(y_test, y_pred_rf)

acc_svm = accuracy_score(y_test, y_pred_svm)
prec_svm = precision_score(y_test, y_pred_svm)

print("="*50)
print("RANDOM FOREST")
print(f"Accuracy: {acc_rf:.4f}")
print(f"Precision: {prec_rf:.4f}")
print(classification_report(y_test, y_pred_rf))

print("="*50)
print("SVM")
print(f"Accuracy: {acc_svm:.4f}")
print(f"Precision: {prec_svm:.4f}")
print(classification_report(y_test, y_pred_svm))
```

8. Guarde sus modelos en formato “pkl” usando joblib, como se realizó en clase.

```python
import joblib

# Guardar modelos y escalador
joblib.dump(modelo_rf, 'modelo_rf_heart.pkl')
joblib.dump(modelo_svm, 'modelo_svm_heart.pkl')
joblib.dump(scaler, 'scaler_heart.pkl')

print("Modelos guardados con jolib con .pkl")
```
